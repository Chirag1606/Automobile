```python
#Car maker Percentage 

make_counts = df['make'].value_counts()
plt.pie(make_counts, autopct= "%.1f%%", labels= make_counts.index);
```

```python
#Fuel type Comparison in cars
fuel_type_counts = df['fuel_type'].value_counts()
plt.pie(fuel_type_counts, autopct= "%.1f%%", labels= fuel_type_counts.index);
```

```python
#Fuel system % in cars
fuel_system_counts = df['fuel_system'].value_counts()
plt.pie(fuel_system_counts, autopct= "%.1f%%", labels= fuel_system_counts.index);
```

```python
#Body Style Comparison
body_style_counts = df['body_style'].value_counts()
plt.pie(body_style_counts, autopct= "%.1f%%", labels= body_style_counts.index);
```

```python
#No of doors %
door_counts = df['number_of_doors'].value_counts()
plt.pie(door_counts, autopct= "%.1f%%", labels= door_counts.index);
```

```python
#Drive Wheel Counts
drive_wheels_counts = df['drive_wheels'].value_counts()
plt.pie(drive_wheels_counts, autopct= "%.1f%%", labels= drive_wheels_counts.index);
```

```python
#Engine location Comparison in cars
engine_location_counts = df['engine_location'].value_counts()
plt.pie(engine_location_counts, autopct= "%.1f%%", labels= engine_location_counts.index);
```

```python
#ENgine type comparion in cars
engine_type_counts = df['engine_type'].value_counts()
plt.pie(engine_type_counts, autopct= "%.1f%%", labels= engine_type_counts.index);
```
